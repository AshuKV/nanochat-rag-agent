# Task 1: RAG Query Agent Demo

This notebook demonstrates the RAG Query Agent for answering technical questions about the NanoChat codebase.

## Features
- Accepts technical questions about the codebase
- Retrieves relevant context from vector database
- Generates coherent, well-structured answers
- Provides source file traceability
- Can be used as a standalone tool or integrated into LangGraph workflows


## Setup and Imports


```python
from rag_query_agent import RAGQueryAgent, rag_tool
import json
```

## Initialize the RAG Query Agent

Make sure your LLM server (e.g., LM Studio) is running on `http://localhost:1234/v1`


```python
# Initialize the agent
agent = RAGQueryAgent(
    base_url="http://localhost:1234/v1",
    api_key="lm_studio",
    temperature=0.0,
    max_tokens=10000,
    k_retrieval=8,
    device="cpu"
)

print("RAG Query Agent initialized successfully!")
```

## Example 1: Query the Agent


```python
question = "What is the main purpose of the NanoChat codebase?"

result = agent.query(question)

print("="*80)
print("QUESTION:")
print("="*80)
print(result["question"])
print("\n" + "="*80)
print("ANSWER:")
print("="*80)
print(result["answer"])
print("\n" + "="*80)
print("SOURCES:")
print("="*80)
for i, source in enumerate(result["sources"], 1):
    print(f"{i}. {source}")
```
